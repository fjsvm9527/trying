```python
import torch
import torch.nn as nn
from torchviz import make_dot

# 
class InceptionModule(nn.Module):
    def __init__(self,input_size,model_param):
        in_channels = input_size
        out_channels_per_path = model_param['out_channels_per_path']
        super().__init__()
        self.conv1 = nn.Conv1d(in_channels, out_channels_per_path, kernel_size=1, padding='same')
        self.conv3 = nn.Conv1d(in_channels, out_channels_per_path, kernel_size=3, padding='same')
        self.conv5 = nn.Conv1d(in_channels, out_channels_per_path, kernel_size=5, padding='same')
        self.pool = nn.MaxPool1d(kernel_size=3, stride=1, padding=1)
        self.pool_conv = nn.Conv1d(in_channels, out_channels_per_path, kernel_size=1, padding='same')

    def forward(self, x):
        x_permuted = x.permute(0, 2, 1)
        out1 = self.conv1(x_permuted)
        out3 = self.conv3(x_permuted)
        out5 = self.conv5(x_permuted)
        pool_out = self.pool(x_permuted)
        out_pool = self.pool_conv(pool_out)
        concatenated = torch.cat([out1, out3, out5, out_pool], dim=1)
        return concatenated.permute(0, 2, 1)
class MLP(nn.Module):
    def __init__(self, param):
        """
         MLP 

        :
        - self.dims (list):  [input_dim, hidden_dim1, hidden_dim2, ..., output_dim]
        - use_bn (bool):  Batch Normalization
        - use_leakyrelu (bool):  LeakyReLU 
        - dropout_rate (float): Dropout 0.0  Dropout
        """
        super(MLP, self).__init__()
        self.layers = nn.ModuleList()
        self.use_bn = param['use_bn']
        self.use_leakyrelu = param['use_leakyrelu']
        self.dropout_rate = param['dropout_rate']
        self.dims = param['dims']

        # 
        for i in range(len(self.dims) - 1):
            self.layers.append(nn.Linear(self.dims[i], self.dims[i + 1]))
            if i < len(self.dims) - 2:  # 
                if self.use_bn:
                    self.layers.append(nn.BatchNorm1d(self.dims[i + 1]))
                if self.use_leakyrelu:
                    self.layers.append(nn.ReLU())
                if self.dropout_rate > 0.0:
                    self.layers.append(nn.Dropout(self.dropout_rate))

    def forward(self, x):
        for layer in self.layers:
            x = layer(x)
        return x
class Inception_LSTM(nn.Module):
    def __init__(self, num_features,model_param):
        super().__init__()
        self.inception = InceptionModule(num_features, model_param['Inception_param'])
        gru_input_size = model_param['Inception_param']['out_channels_per_path'] * 4
        self.lstm = nn.LSTM(
            input_size=gru_input_size,
            hidden_size=model_param['hidden_size'],
            num_layers=model_param['num_layers'],
            batch_first=model_param['batch_first'],
            dropout=model_param['dropout']
        )
        self.mlp = MLP(model_param['MLP_param'])

    def forward(self, x):
        x = self.inception(x)
        lstm_out, _ = self.lstm(x)
        last_time_step_out = lstm_out[:, -1, :]
        prediction = self.mlp(last_time_step_out)
        return prediction.squeeze(-1)

# # 
model = Inception_LSTM(num_features=27,model_param={
      "Inception_param":{
        "out_channels_per_path":16
      },
      "hidden_size":64,
      "num_layers":2,
      "batch_first":True,
      "dropout":0.2,
      "MLP_param":{
        "use_bn":False,
        "use_leakyrelu":True,
        "dropout_rate":0.3,
        "dims":[64,32,1]
      }
    })
# model = MLP({
#         "use_bn":False,
#         "use_leakyrelu":False,
#         "dropout_rate":0.0,
#         "dims":[64,32,1]
#       })
dummy_input = torch.randn(1024,8,27)
output = model(dummy_input)

torch.onnx.export(model,
                  dummy_input,
                  "my_model.onnx",
                  verbose=True,
                  input_names=['input'],
                  output_names=['output'])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
from matplotlib.font_manager import FontProperties
import matplotlib
import matplotlib.pyplot as plt
import matplotlib

# 
plt.rcParams.update({
    'font.family': 'Microsoft YaHei',  # 
    'font.sans-serif': ['Microsoft YaHei', 'SimHei', 'Arial Unicode MS'],
    'axes.unicode_minus': False,
    
    # Arial
    'svg.fonttype': 'none',
    'pdf.fonttype': 42,  # 42=Type3, PDF
    'ps.fonttype': 42,
    'figure.max_open_warning': 0  # 
})


def calculate_and_plot_performance(df: pd.DataFrame):
    """
    Calculates key performance metrics for a trading strategy and plots the results.

    Args:
        df (pd.DataFrame): DataFrame with 'date' and 'daily_return' columns.
                           'date' should be convertible to datetime.
                           'daily_return' should be arithmetic daily returns.
    """
    # --- 1. Data Preparation ---
    if not isinstance(df, pd.DataFrame) or not {'date', 'daily_return'}.issubset(df.columns):
        raise ValueError("Input must be a DataFrame with 'date' and 'daily_return' columns.")

    # Copy the dataframe to avoid modifying the original
    strategy_df = df.copy()

    # Convert 'date' column to datetime and set it as the index
    strategy_df['date'] = pd.to_datetime(strategy_df['date'])
    strategy_df = strategy_df.sort_values(by='date').set_index('date')

    # --- 2. Performance Metrics Calculation ---

    # Trading days per year (standard assumption)
    TRADING_DAYS_PER_YEAR = 252

    # Cumulative Return (as requested, by summing daily returns)
    strategy_df['cumulative_ret'] = strategy_df['daily_return'].cumsum()

    # Total Cumulative Return
    total_return = strategy_df['cumulative_ret'].iloc[-1]

    # Annualized Return
    num_years = len(strategy_df) / TRADING_DAYS_PER_YEAR
    annualized_return = total_return / num_years

    # Annualized Volatility
    annualized_volatility = strategy_df['daily_return'].std() * np.sqrt(TRADING_DAYS_PER_YEAR)

    # Sharpe Ratio (assuming risk-free rate is 0)
    sharpe_ratio = annualized_return / annualized_volatility if annualized_volatility != 0 else 0

    # Drawdown Calculation
    # High Water Mark: The running maximum of the cumulative return
    strategy_df['high_water_mark'] = strategy_df['cumulative_ret'].cummax()
    # Drawdown: The difference between the High Water Mark and the current cumulative return
    strategy_df['drawdown'] = strategy_df['high_water_mark'] - strategy_df['cumulative_ret']
    # Maximum Drawdown
    max_drawdown = strategy_df['drawdown'].max()

    # Calmar Ratio
    calmar_ratio = annualized_return / max_drawdown if max_drawdown != 0 else 0
    
    # --- 3. Print Metrics ---
    print("--- Strategy Performance Metrics ---")
    print(f"Time Period: {strategy_df.index.min().strftime('%Y-%m-%d')} to {strategy_df.index.max().strftime('%Y-%m-%d')}")
    print(f"Total Cumulative Return: {total_return:.2%}")
    print(f"Annualized Return: {annualized_return:.2%}")
    print(f"Annualized Volatility: {annualized_volatility:.2%}")
    print(f"Sharpe Ratio (Rf=0%): {sharpe_ratio:.2f}")
    print(f"Maximum Drawdown (MDD): {max_drawdown:.2%}")
    print(f"Calmar Ratio: {calmar_ratio:.2f}")
    print("-" * 34)


    # --- 4. Plotting ---
    plt.style.use('seaborn-v0_8-whitegrid')
    fig, (ax1, ax2) = plt.subplots(
        nrows=2, 
        ncols=1, 
        sharex=True, 
        figsize=(12, 8), 
        gridspec_kw={'height_ratios': [3, 1]} # Give more space to the main plot
    )
    fig.suptitle('', fontsize=16, y=0.95)

    # Plot 1: Cumulative Return (Equity Curve)
    ax1.plot(strategy_df.index, strategy_df['cumulative_ret'], color='darkblue', linewidth=2)
    ax1.set_ylabel('', fontsize=12)
    ax1.set_title('', fontsize=14)
    # Format Y-axis as percentage
    ax1.yaxis.set_major_formatter(mticker.PercentFormatter(xmax=1.0))
    
    # Plot 2: Drawdown
    ax2.fill_between(strategy_df.index, -strategy_df['drawdown'], 0, color='red', alpha=0.3)
    ax2.plot(strategy_df.index, -strategy_df['drawdown'], color='red', linewidth=1) # Add a line for clarity
    ax2.set_ylabel('Drawdown', fontsize=12)
    ax2.set_title('Drawdown Curve', fontsize=14)
    # Format Y-axis as percentage
    ax2.yaxis.set_major_formatter(mticker.PercentFormatter(xmax=1.0))
    
    # Common formatting
    ax2.set_xlabel('Date', fontsize=12)
    fig.tight_layout(rect=[0, 0, 1, 0.93]) # Adjust layout to make space for suptitle
    
    # Add metrics text box to the plot for a quick summary
    stats_text = (
        f": {annualized_return:.2%}\n"
        f": {annualized_volatility:.2%}\n"
        f": {sharpe_ratio:.2f}\n"
        f": {max_drawdown:.2%}\n"
    )
    ax1.text(0.02, 0.98, stats_text, transform=ax1.transAxes, fontsize=10,
             verticalalignment='top', bbox=dict(boxstyle='round,pad=0.5', facecolor='white', alpha=0.7))

    plt.show()

    return strategy_df


# --- Example Usage ---
# Create a sample DataFrame. REPLACE THIS WITH YOUR DATAFRAME.
np.random.seed(42) # for reproducibility
import pandas as pd
results = pd.read_csv("Total_res.csv")

# Run the analysis and plotting function
performance_summary = calculate_and_plot_performance(results)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker

def predict_ewma_volatility(df: pd.DataFrame, span: int = 10):
    """
     (EWMA) 

    Args:
        df (pd.DataFrame):  'daily_vol'  DataFrame
                            datetime 
        span (int):  EWMA 

    Returns:
        tuple: (,  DataFrame)
    """
    if not isinstance(df.index, pd.DatetimeIndex):
        raise ValueError("DataFrame index must be a DatetimeIndex.")
    
    df_ewma = df.copy()

    # --- 1.  EWMA  ---
    # .ewm() 
    # span=10 10
    # adjust=False 
    ewma_series = df_ewma['daily_vol'].ewm(span=span, adjust=False).mean()

    # --- 2.  ---
    #  t  EWMA  t+1 
    #  .shift(1) 
    df_ewma['predicted_vol'] = ewma_series.shift(1)

    # --- 3.  ---
    #  EWMA 
    next_day_prediction = ewma_series.iloc[-1]
    
    print(f"--- EWMA (span={span})  ---")
    print(f"\n: {next_day_prediction:.6f}")
    
    # --- 4.  ---
    plt.style.use('seaborn-v0_8-whitegrid')
    fig, ax = plt.subplots(figsize=(14, 7))
    
    ax.plot(df_ewma.index, df_ewma['daily_vol'], label='Actual Daily Volatility', color='skyblue', linewidth=1.5)
    ax.plot(df_ewma.index, df_ewma['predicted_vol'], label=f'EWMA Prediction (span={span})', color='coral', linestyle='--', linewidth=2)
    
    ax.set_title('EWMA Volatility Prediction Model', fontsize=16)
    ax.set_ylabel('Volatility', fontsize=12)
    ax.set_xlabel('Date', fontsize=12)
    ax.legend(fontsize=12)
    ax.grid(True, which='both', linestyle='--', linewidth=0.5)
    
    # 
    plt.setp(ax.get_xticklabels(), rotation=45, ha="right")
    fig.tight_layout()
    plt.show()
    
    return next_day_prediction, df_ewma

_,df_ewa = predict_ewma_volatility(date_performance,span=10)
```

```python
import pandas as pd

def calculate_trading_num(df):
    """
    

    :
    df (pd.DataFrame): DataFrame
                        'date'  'position' 

    :
    pd.DataFrame: DataFrame 'date'  'trading_num' 
    """
    # 
    df = df.sort_values(by=['date', 'time']).copy()

    # 
    df['position_diff'] = df.groupby('date')['position'].diff().fillna(0)

    # abs(position_diff)
    # 0 -> 1  or 0 -> -1: abs(1) = 1, abs(-1) = 1
    # 1 -> -1 or -1 -> 1: abs(-2) = 2, abs(2) = 2
    df['trades'] = df['position_diff'].abs()

    # 
    daily_trades = df.groupby('date')['trades'].sum().reset_index()
    daily_trades = daily_trades.rename(columns={'trades': 'trading_num'})
    daily_trades['trading_num'] = daily_trades['trading_num'] / 2

    return daily_trades

# --------  --------
df = pd.read_csv("D:\JT_Summer\Total_signal.csv")
# 
daily_trading_num = calculate_trading_num(df)

# 
print("\n:")
print(daily_trading_num)
```
